###Wyświetlanie obrazów przy użyciu Matplotlib

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('office.jpg')
```

```python
img.shape
```

```python
plt.imshow(img)
_ = plt.axis('off')
```

```python
plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.axis('off')
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.axis('off')
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.axis('off')
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
plt.axis('off')
```

```python
plt.figure(figsize=(12, 10))

for idx, image in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(220 + idx + 1)
    plt.imshow(plt.imread(image))
    plt.axis('off')
```

```python
plt.figure(figsize=(12, 10))

for idx, image in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
    plt.subplot(140 + idx + 1)
    plt.imshow(plt.imread(image))
    plt.axis('off')
```
